# Export UK-KingsCollege site temperature
- This script is used to export sWBGT at the UK-KingsCollege site on 2013-07-22;
- Simulations: CNTL, BASE;

```python
import xarray as xr
import pandas as pd
import numpy as np
```

```python
GRIDNAME='UK-Kin'
name_lcz = GRIDNAME + '_base_landunit'
name_def = GRIDNAME + '_def_landunit'
path = '/mnt/iusers01/fatpou01/sees01/a16404ys/scratch/Projects/archive/0project2/0sp/'
f_lcz = path + name_lcz + '/lnd/hist/'
f_def = path + name_def + '/lnd/hist/'
var = ['SWBGT_U']
start_time = '2013-07-22T00:00:00'
end_time = '2013-07-22T23:30:00'
mid_date = '2013-01-01T00:00:00'
filename_lcz = name_lcz + '.clm2.h0.2002-01-01-00000.nc'
filename_lcz_landunit = name_lcz + '.clm2.h1.2002-01-01-00000.nc'

filename_def = name_def + '.clm2.h0.2002-01-01-00000.nc'
ds_lcz = xr.open_dataset(f_lcz + filename_lcz).sel(time=slice(start_time, end_time))
ds_lcz_landunit = xr.open_dataset(f_lcz + filename_lcz_landunit).sel(time=slice(start_time, end_time))

ds_def = xr.open_dataset(f_def + filename_def).sel(time=slice(start_time, end_time))
ds_lcz['time']=ds_lcz['time'].dt.round('min')
ds_lcz['time']=ds_lcz['time'].dt.ceil('min')
ds_def['time']=ds_def['time'].dt.round('min')
ds_def['time']=ds_def['time'].dt.ceil('min')
tag = ['_def', '_lcz']
```
